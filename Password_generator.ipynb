```python
import random
import string
import secrets

def generate_password(length=12, use_uppercase=True, use_lowercase=True, 
                      use_digits=True, use_symbols=True, exclude_ambiguous=False):
    """
    Generate a random, complex password.
    
    Args:
        length (int): Desired password length (default: 12)
        use_uppercase (bool): Include uppercase letters
        use_lowercase (bool): Include lowercase letters
        use_digits (bool): Include digits
        use_symbols (bool): Include special characters
        exclude_ambiguous (bool): Exclude ambiguous characters (il1Lo0O)
    
    Returns:
        str: Generated password
    """
    # Define character sets
    lowercase = 'abcdefghijklmnopqrstuvwxyz'
    uppercase = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    digits = '0123456789'
    symbols = '!@#$%^&*()_+-=[]{}|;:,.<>?'
    
    # Remove ambiguous characters if requested
    if exclude_ambiguous:
        ambiguous = 'il1Lo0O'
        lowercase = ''.join(c for c in lowercase if c not in ambiguous)
        uppercase = ''.join(c for c in uppercase if c not in ambiguous)
        digits = ''.join(c for c in digits if c not in ambiguous)
    
    # Build the character pool
    char_pool = ''
    if use_lowercase:
        char_pool += lowercase
    if use_uppercase:
        char_pool += uppercase
    if use_digits:
        char_pool += digits
    if use_symbols:
        char_pool += symbols
    
    # Ensure at least one character type is selected
    if not char_pool:
        raise ValueError("At least one character type must be selected!")
    
    # Generate password using secrets module for cryptographically secure randomness
    password = ''.join(secrets.choice(char_pool) for _ in range(length))
    
    return password

def generate_password_with_requirements(length=12):
    """
    Generate a password that includes at least one character from each type.
    This ensures the password meets complexity requirements.
    """
    # Define character sets
    lowercase = 'abcdefghijklmnopqrstuvwxyz'
    uppercase = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    digits = '0123456789'
    symbols = '!@#$%^&*()_+-=[]{}|;:,.<>?'
    
    # Ensure minimum length of 4 for all character types
    if length < 4:
        print("⚠️  Password length too short. Setting to minimum 4 characters.")
        length = 4
    
    # Generate at least one of each type
    password = [
        secrets.choice(lowercase),
        secrets.choice(uppercase),
        secrets.choice(digits),
        secrets.choice(symbols)
    ]
    
    # Fill the rest with random characters from all types
    all_chars = lowercase + uppercase + digits + symbols
    for _ in range(length - 4):
        password.append(secrets.choice(all_chars))
    
    # Shuffle the password to avoid predictable pattern
    random.shuffle(password)
    
    return ''.join(password)

def check_password_strength(password):
    """
    Check the strength of a password based on various criteria.
    """
    score = 0
    feedback = []
    
    # Length check
    if len(password) >= 12:
        score += 2
        feedback.append("✅ Good length (12+ characters)")
    elif len(password) >= 8:
        score += 1
        feedback.append("⚠️  Moderate length (8-11 characters)")
    else:
        feedback.append("❌ Too short (less than 8 characters)")
    
    # Character type checks
    has_lower = any(c.islower() for c in password)
    has_upper = any(c.isupper() for c in password)
    has_digit = any(c.isdigit() for c in password)
    has_symbol = any(c in '!@#$%^&*()_+-=[]{}|;:,.<>?' for c in password)
    
    if has_lower:
        score += 1
        feedback.append("✅ Contains lowercase letters")
    else:
        feedback.append("❌ Missing lowercase letters")
    
    if has_upper:
        score += 1
        feedback.append("✅ Contains uppercase letters")
    else:
        feedback.append("❌ Missing uppercase letters")
    
    if has_digit:
        score += 1
        feedback.append("✅ Contains digits")
    else:
        feedback.append("❌ Missing digits")
    
    if has_symbol:
        score += 1
        feedback.append("✅ Contains special characters")
    else:
        feedback.append("⚠️  No special characters")
    
    # Calculate strength rating
    if score >= 6:
        strength = "💪 Strong"
    elif score >= 4:
        strength = "👍 Moderate"
    else:
        strength = "🔴 Weak"
    
    return strength, feedback

def password_generator_tool():
    """
    Interactive password generator tool with user input.
    """
    print("\n" + "=" * 60)
    print("        🔐 PASSWORD GENERATOR")
    print("=" * 60)
    
    while True:
        try:
            # Get password length
            print("\n📏 Enter password length:")
            length_input = input("   (minimum 4, default 12): ").strip()
            
            if length_input == "":
                length = 12
            else:
                length = int(length_input)
                if length < 4:
                    print("⚠️  Minimum length is 4. Using 4 characters.")
                    length = 4
                elif length > 128:
                    print("⚠️  Maximum recommended length is 128. Using 128.")
                    length = 128
            
            # Ask for complexity preferences
            print("\n⚙️  Complexity Options:")
            print("   Include uppercase letters? (y/n, default: y)")
            use_upper = input("   ").strip().lower() != 'n'
            
            print("   Include lowercase letters? (y/n, default: y)")
            use_lower = input("   ").strip().lower() != 'n'
            
            print("   Include numbers? (y/n, default: y)")
            use_digits = input("   ").strip().lower() != 'n'
            
            print("   Include symbols? (y/n, default: y)")
            use_symbols = input("   ").strip().lower() != 'n'
            
            print("   Exclude ambiguous characters (il1Lo0O)? (y/n, default: n)")
            exclude_ambiguous = input("   ").strip().lower() == 'y'
            
            print("\n🔄 Generating password...")
            print("-" * 60)
            
            # Generate multiple passwords for user to choose from
            print("\n🔑 Here are 5 generated passwords:")
            passwords = []
            for i in range(1, 6):
                try:
                    pwd = generate_password(
                        length=length,
                        use_uppercase=use_upper,
                        use_lowercase=use_lower,
                        use_digits=use_digits,
                        use_symbols=use_symbols,
                        exclude_ambiguous=exclude_ambiguous
                    )
                    passwords.append(pwd)
                    print(f"   {i}. {pwd}")
                except ValueError as e:
                    print(f"   {i}. ❌ Error: {e}")
                    return
            
            # Also generate a guaranteed-complex password
            print("\n✨ Strong password (guaranteed to include all types):")
            strong_password = generate_password_with_requirements(length)
            print(f"   🔒 {strong_password}")
            
            # Show strength analysis for the strong password
            print("\n📊 Password Strength Analysis:")
            strength, feedback = check_password_strength(strong_password)
            print(f"   {strength}")
            for item in feedback:
                print(f"   {item}")
            
            # Ask if user wants to save the password
            print("\n💾 Options:")
            print("   Save the strong password to a file? (y/n)")
            save_choice = input("   ").strip().lower()
            
            if save_choice == 'y':
                filename = "generated_password.txt"
                with open(filename, 'w') as f:
                    f.write(f"Generated Password: {strong_password}\n")
                    f.write(f"Length: {length} characters\n")
                    f.write(f"Generated on: {__import__('datetime').datetime.now()}\n")
                print(f"✅ Password saved to '{filename}'")
            
            # Ask if user wants to generate another
            print("\n🔄 Generate another password? (y/n)")
            again = input("   ").strip().lower()
            if again != 'y':
                print("\n👋 Thank you for using the Password Generator!")
                break
            
        except ValueError:
            print("❌ Invalid input! Please enter a valid number for password length.")
        except KeyboardInterrupt:
            print("\n\n👋 Goodbye!")
            break
        except Exception as e:
            print(f"❌ An error occurred: {e}")
            break

# Quick generation function for command-line use
def quick_password(length=12, strong=True):
    """
    Quick password generation without the interactive menu.
    
    Args:
        length (int): Password length
        strong (bool): If True, guarantees at least one of each type
    
    Returns:
        str: Generated password
    """
    if strong:
        return generate_password_with_requirements(length)
    else:
        return generate_password(length=length)

if __name__ == "__main__":
    # Run the interactive tool
    password_generator_tool()
    
    # Uncomment below for quick generation example:
    # print("\n🔑 Quick Password Examples:")
    # print(f"Standard: {quick_password(16, False)}")
    # print(f"Strong: {quick_password(16, True)}")
```
