```python
## def expense_tracker():
    """
    A simple expense tracker that allows users to enter expenses
    and displays the total amount spent.
    """
    total_spent = 0.0
    expense_count = 0
    
    print("=" * 40)
    print("     EXPENSE TRACKER")
    print("=" * 40)
    print("Enter expense amounts (e.g., 100, 50, 20)")
    print("Type 'done' or 'quit' to finish")
    print("-" * 40)
    
    while True:
        try:
            # Get user input
            user_input = input("Enter expense amount: ").strip()
            
            # Check if user wants to quit
            if user_input.lower() in ['done', 'quit', 'exit', 'q']:
                break
            
            # Try to convert input to float
            expense = float(user_input)
            
            # Validate that expense is positive
            if expense < 0:
                print("❌ Please enter a positive amount.")
                continue
            
            # Add expense to total
            total_spent += expense
            expense_count += 1
            
            # Show current total
            print(f"✅ Added: ${expense:.2f}")
            print(f"   Current total: ${total_spent:.2f}")
            print("-" * 40)
            
        except ValueError:
            print("❌ Invalid input. Please enter a number or type 'done' to finish.")
    
    # Display final summary
    print("\n" + "=" * 40)
    print("     EXPENSE SUMMARY")
    print("=" * 40)
    
    if expense_count == 0:
        print("No expenses were entered.")
    else:
        print(f"Total number of expenses: {expense_count}")
        print(f"Total amount spent: ${total_spent:.2f}")
        if expense_count > 0:
            print(f"Average expense: ${total_spent / expense_count:.2f}")
    
    print("=" * 40)
    print("Thank you for using Expense Tracker!")

# Alternative version: Simple one-time input version
def simple_expense_tracker():
    """
    A simpler version where users can enter multiple expenses
    and see the total at the end.
    """
    expenses = []
    
    print("\n" + "=" * 40)
    print("     SIMPLE EXPENSE TRACKER")
    print("=" * 40)
    print("Enter expenses one by one.")
    print("Press Enter without typing anything to finish.")
    print("-" * 40)
    
    while True:
        user_input = input("Expense amount: ").strip()
        
        if user_input == "":
            break
        
        try:
            expense = float(user_input)
            if expense < 0:
                print("❌ Please enter a positive amount.")
                continue
            expenses.append(expense)
            print(f"   Added: ${expense:.2f}")
        except ValueError:
            print("❌ Invalid input. Please enter a number.")
    
    # Show results
    if expenses:
        total = sum(expenses)
        print("\n" + "=" * 40)
        print("     EXPENSE SUMMARY")
        print("=" * 40)
        print(f"Number of expenses: {len(expenses)}")
        print(f"Total spent: ${total:.2f}")
        print(f"Individual expenses: {[f'${e:.2f}' for e in expenses]}")
        print("=" * 40)
    else:
        print("\nNo expenses were entered.")

# Run the main program
if __name__ == "__main__":
    # Choose which version to run
    print("Choose expense tracker version:")
    print("1. Interactive (with running total)")
    print("2. Simple (show total at end)")
    
    choice = input("Enter your choice (1 or 2): ").strip()
    
    if choice == "1":
        expense_tracker()
    elif choice == "2":
        simple_expense_tracker()
    else:
        print("Invalid choice. Running the interactive version...")
        expense_tracker()
```
