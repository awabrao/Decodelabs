```python
def show_menu():
    print("\n--- Task Manager ---")
    print("1. Add task")
    print("2. View tasks")
    print("3. Exit")

def add_task(tasks):
    task = input("Enter task: ").strip()
    if task:
        tasks.append(task)
        print(f"Added: {task}")
    else:
        print("Task cannot be empty.")

def view_tasks(tasks):
    if not tasks:
        print("No tasks yet.")
        return

    print("\nYour tasks:")
    for i, task in enumerate(tasks, start=1):
        print(f"  {i}. {task}")

def main():
    tasks = []

    while True:
        show_menu()
        choice = input("Choose an option (1-3): ").strip()

        if choice == "1":
            add_task(tasks)
        elif choice == "2":
            view_tasks(tasks)
        elif choice == "3":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

if __name__ == "__main__":
    main()
```
